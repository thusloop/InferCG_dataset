主要针对pre_knowledge

```python
import json
# 加载 JSON 文件
def load_json(filepath):
    with open(filepath, 'r', encoding='utf-8') as f:
        return json.load(f)

def solve(name, clean_pre):
    pre_annotations_path = "../STAR/pre_knowledge/{}_pre_annotations.json".format(name)
    import_info_path = "../STAR/pre_knowledge/{}_import_info.json".format(name)
    inherited_info_path = "../STAR/pre_knowledge/{}_pre_inherited.json".format(name)

    pre_annotations_data = load_json(pre_annotations_path)
    import_info = load_json(import_info_path)
    inherited_info = load_json(inherited_info_path)

    pre_annotations_data_new = {}
    import_info_new = {}
    inherited_info_new = {}



    for key , val in pre_annotations_data.items():
        if key.split('.')[0] == clean_pre:
            key = '.'.join(key.split('.')[1:])
        if val['API_name'].split('.')[0] == clean_pre:
            val['API_name'] = '.'.join(val['API_name'].split('.')[1:])
        if val['loc_name'].split('.')[0] == clean_pre:
            val['loc_name'] = '.'.join(val['loc_name'].split('.')[1:])
        if val['filepath'].split('.')[0] == clean_pre:
            val['filepath'] = '.'.join(val['filepath'].split('.')[1:])
            
        pre_annotations_data_new[key] = val

    for key , val in import_info.items():
        if key.split('.')[0] == clean_pre:
            key = '.'.join(key.split('.')[1:])
        import_info_new[key] = val

    for key , val in inherited_info.items():
        if key.split('.')[0] == clean_pre:
            key = '.'.join(key.split('.')[1:])
        val['inherited'] = [
            '.'.join(item.split('.')[1:]) if item.split('.')[0] == clean_pre else item
            for item in val['inherited']
        ]
        inherited_info_new[key] = val

    with open(pre_annotations_path, "w", encoding="utf-8") as file:
        json.dump(pre_annotations_data_new, file, indent=4, ensure_ascii=False)
    with open(import_info_path, "w", encoding="utf-8") as file:
        json.dump(import_info_new, file, indent=4, ensure_ascii=False)
    with open(inherited_info_path, "w", encoding="utf-8") as file:
        json.dump(inherited_info_new, file, indent=4, ensure_ascii=False)
```
